## JSON example, with string

+ demonstrates creation of normalized dataframes (tables) from nested json string
+ source: http://pandas.pydata.org/pandas-docs/stable/io.html#normalization

```python
# define json string
data = [{'state': 'Florida', 
         'shortname': 'FL',
         'info': {'governor': 'Rick Scott'},
         'counties': [{'name': 'Dade', 'population': 12345},
                      {'name': 'Broward', 'population': 40000},
                      {'name': 'Palm Beach', 'population': 60000}]},
        {'state': 'Ohio',
         'shortname': 'OH',
         'info': {'governor': 'John Kasich'},
         'counties': [{'name': 'Summit', 'population': 1234},
                      {'name': 'Cuyahoga', 'population': 1337}]}]
```

```python
# use normalization to create tables from nested element
json_normalize(data, 'counties')
```

```python
# further populate tables created from nested element
json_normalize(data, 'counties', ['state', 'shortname', ['info', 'governor']])
```

## JSON exercise

Using data in file 'data/world_bank_projects.json' and the techniques demonstrated above,
1. Find the 10 countries with most projects
2. Find the top 10 major project themes (using column 'mjtheme_namecode')
3. In 2. above you will notice that some entries have only the code and the name is missing. Create a dataframe with the missing names filled in.

```python
import pandas as pd
import json
from pandas.io.json import json_normalize

# Load data to explore
projects_json = json.load((open('data/world_bank_projects.json')))
```

```python
len(projects_json)
```

```python
# 1. Find the 10 countries with most projects

# Use a counter dict to count how many times each country is present (how many projects they have)
from collections import Counter
countries = Counter(item['countryshortname'] for item in projects_json)

print("The 10 most countries with most projects are the following:")
print(", ".join([i[0] for i in countries.most_common(10)]))
```

```python
# 2. Find the top 10 major project themes (using column 'mjtheme_namecode')

theme_names = json_normalize(projects_json, 'mjtheme_namecode')
# With the option below you can get rid of the empty 'names'. Will be fixed in a better way in next exercise.
theme_names = theme_names[theme_names['name'] != '']
theme_names = theme_names['name'].value_counts()

# Printing this result so we can compare with the filled in values later
theme_names
```

```python
print("The top 10 major project themes are:")
print("\n ".join(theme_names.index.to_list()[:10]))
```

```python
# 3. In 2. above you will notice that some entries have only the code and the name is missing. 
# Create a dataframe with the missing names filled in.

# I'll replace the empty strings with NaN and then use fillna to backfill from the next valid name
import numpy as np
theme_names = json_normalize(projects_json, 'mjtheme_namecode')

# First sort by code and then name, so we can backfill safely
sort_themes = theme_names.sort_values(['code', 'name'])
filled = sort_themes.replace('', np.nan)
f = filled.fillna(method='bfill')
f
```

```python
# We can see that the counts change, of course. But the top 10 themes are the same.
f['name'].value_counts()
```

```python
# Plotting 
f['name'].value_counts().plot(kind='barh')
```
